```python
# Libraries needed
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

<h1>Script</h1>

```python
def insta_like(image,instafilter):
    blur_image = cv2.blur(img,(5,5))
    
    b = blur_image[:,:,0]
    g = blur_image[:,:,1]
    r = blur_image[:,:,2]
     
    if instafilter == 'Inkwell':
        gray = cv2.cvtColor(blur_image,cv2.COLOR_RGB2GRAY)
        return gray
    elif instafilter == 'Lily':
        b1=b*0.5
        g1=g*1
        r1=r*0.1
        b_ = np.ceil(b1).astype(np.uint8) 
        g_ = np.ceil(g1).astype(np.uint8) 
        r_ = np.ceil(r1).astype(np.uint8)
        bgr =cv2.merge((b_,g_,r_))
        alpha = 1
        beta = 30
        result = cv2.addWeighted(bgr,alpha,np.zeros(bgr.shape,bgr.dtype),0,beta)
        return result
    elif instafilter == 'Poprocket':
        b1=b*1
        g1=g*0
        r1=r*1
        b_ = np.ceil(b1).astype(np.uint8) 
        g_ = np.ceil(g1).astype(np.uint8) 
        r_ = np.ceil(r1).astype(np.uint8)
        bgr =cv2.merge((b_,g_,r_))
        alpha = 1
        beta = 40
        result = cv2.addWeighted(bgr,alpha,np.zeros(bgr.shape,bgr.dtype),0,beta)
        return result
    elif instafilter == 'LordKelvin':
        b1=b*0
        g1=g*1
        r1=r*1
        b_ = np.ceil(b1).astype(np.uint8) 
        g_ = np.ceil(g1).astype(np.uint8) 
        r_ = np.ceil(r1).astype(np.uint8)
        bgr =cv2.merge((b_,g_,r_))
        alpha = 1
        beta = 40
        result = cv2.addWeighted(bgr,alpha,np.zeros(bgr.shape,bgr.dtype),0,beta)
        return result
    elif instafilter == 'Gotham':
        b1=b*1
        g1=g*0.5
        r1=r*0.5
        b_ = np.ceil(b1).astype(np.uint8) 
        g_ = np.ceil(g1).astype(np.uint8) 
        r_ = np.ceil(r1).astype(np.uint8)
        bgr =cv2.merge((b_,g_,r_))
        alpha = 1
        beta = 20
        result = cv2.addWeighted(bgr,alpha,np.zeros(bgr.shape,bgr.dtype),0,beta)
        return result
    else:
        return blur_image
```

<h1>Printing image using Matplotlib</h>
<h3>This function converts bgr image to rgb in order to give the same output for both matplotlib and opencv</h3>

```python
def bgr_to_rgb(image):
    im2 = image.copy()
    im2[:, :, 0] = image[:, :, 2]
    im2[:, :, 2] = image[:, :, 0]
    plt.imshow(im2)
```

<h1>Input Image</h1>

```python
img = cv2.imread('input.jpg',1)
bgr_to_rgb(img)
#cv2.imshow('output.jpg',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

<h1>Inkwell Filter</h1>

```python
inkwell_img = insta_like(img,'Inkwell')
plt.imshow(inkwell_img, cmap='gray')
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
lily_filter = insta_like(img,'Lily')
bgr_to_rgb(lily_filter)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
lordkelvin_filter = insta_like(img,'LordKelvin')
bgr_to_rgb(lordkelvin_filter)
cv2.waitKey()
cv2.destroyAllWindows()
```
